```python
import requests
import matplotlib.pyplot as plt
import time
```

```python
from_time = 1698638400
to_time = 1698677288
```

```python
start_time = time.perf_counter()
res = requests.get(f'http://localhost:8000/get_time_series?from_time={from_time}&to_time={to_time}&chart_width=500&chart_height=100')
res = res.json()
plt.plot([r[0] for r in res], [r[1] for r in res])
end_time = time.perf_counter()
elapsed_time = end_time - start_time
print(elapsed_time)
```

```python
start_time = time.perf_counter()
raw_res = requests.get(f'http://localhost:8000/get_raw_time_series?from_time={from_time}&to_time={to_time}')
raw_res = raw_res.json()
plt.plot([r[0] for r in raw_res], [r[1] for r in raw_res])
end_time = time.perf_counter()
elapsed_time = end_time - start_time
print(elapsed_time)
```
